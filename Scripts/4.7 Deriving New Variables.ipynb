## 03. Define new functions

```python
#sort products into three price categories
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts()
```

### 03.01 loc function

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

### 03.02 Finding the busiest days

```python
df_ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

Reminder that 0 is Saturday

```python
result = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
df_ords_prods_merge['busiest_day'] = result
```

```python
df_ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

## 04 Task

**Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method.**

```python
#No "And/OR" logic is necessary since I'm just editing entries line by line

result_2 = []

for x in df_ords_prods_merge["orders_day_of_week"]:
  if x == 0:
    result_2.append("Busiest days")
  elif x == 1:
    result_2.append("Busiest days")
  elif x == 4:
    result_2.append("Least busy days")      
  elif x == 3:
    result_2.append("Least busy days")
  else:
    result_2.append("Regularly busy")
```

**Check the values of this new column for accuracy. Note any observations in markdown format.**

```python
result_2
```

```python
df_ords_prods_merge['busiest_days'] = result_2
```

```python
df_ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

**Result**: The numbers add up. *Busiest days* is the sum of days 0 and 1 and *Least busy days* is the sum of 3 and 4.
**Observation**: An average might be more meaningful than the total number, since the two calculated categories comprise two days, whereas the *Regularly busy* category comprises three days.
