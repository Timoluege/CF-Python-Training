## 03. Combining Data

```python
df_prods.head()
```

```python
df_orders_products_combined.head()
```

**Based on what I see here I want to perform a merge on product_id but first I want to get rid of the old _merge column.**

I don't need to worry about which type of join I should use as this information was provided. 

```python
#Drop _merge column from df_orders_products_combined and save as df_orders_products_combined_2
df_orders_products_combined_2 = df_orders_products_combined.drop(columns = ['_merge'])
```

```python
df_orders_products_combined_2.head()
```

```python
# Test merge without creating a new df. Compared to V1 of this file I switched order of merge with df_prods first this time.
pd.merge(df_prods,df_orders_products_combined_2, on = ['product_id'], indicator = True)
```

```python
#Merge both datasets on 'product_id' as df_orders_products_id_combined. Compared to V1 of this file I switched order of merge with df_prods first this time.
df_orders_products_id_combined = df_prods.merge(df_orders_products_combined_2, on = 'product_id', indicator = True)
```

```python
df_orders_products_id_combined['_merge'].value_counts()
```

**Observation** We now have 29.630 fewer entries in the database than before. That is a loss of 0.9% of data which I find acceptable. 

## 04. Exporting Data

```python
#Export Data as Pickle
df_orders_products_id_combined.to_pickle(os.path.join(path, 'sub-folder','sub-sub-folder', 'yourfilename.pkl'))
```
