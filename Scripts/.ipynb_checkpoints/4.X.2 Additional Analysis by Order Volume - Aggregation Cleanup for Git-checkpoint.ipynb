## 03. Calculating Order Values

```python
#Create variable order_value summing up all prices for all orders
order_value = df_orders.groupby('order_id')['prices'].sum().reset_index()
```

```python
order_value
```

```python
# Rename the prices column to show that this is the sum of all prices in the order
order_value.rename(columns={'prices': 'sum_order_value'}, inplace=True)
```

```python
order_value
```

```python
# Merge sum_order_value with df as df_order_values
df_order_values = pd.merge(df_orders, order_value, on='order_id', how='left')
```

```python
df_order_values.head()
```

## 04. Analysis

```python
df_order_values.groupby('order_id')
```

```python
#Calculating the mean order value based on region
df_order_values.groupby('region').agg({'sum_order_value': ['mean']})
```

There is no big difference between order values in the different regions.

```python
#Calculating the mean order value based on region
df_order_values.groupby('fam_status').agg({'sum_order_value': ['mean']})
```

There is no big difference between order values between family status.

```python
#Look at fam_status across regions
crosstab = pd.crosstab(df_order_values['fam_status'], df_order_values['region'])
```

```python
crosstab
```

```python
#Calculating based on percentages
crosstab_percentage = pd.crosstab(df_order_values['fam_status'], df_order_values['region'], normalize='columns') * 100
```

```python
crosstab_percentage
```

The distribution of fam_status across the regions is about the same. There is probably nothing to be gained by breaking fam_status down by region.

### 04.01. Visualizations

```python
#Creating a histogram
hist_order_value = df_order_values['sum_order_value'].plot.hist(bins = 25)
```

```python
#Scatterplot to show distribution of order values and identify possible additional issues with the data
sns.scatterplot(x = 'sum_order_value', y = 'sum_order_value',data = df_order_values)
```

```python
df_order_values['sum_order_value'].describe()
```

Creating scatterplot to show income and order volume, including R²-value to see whether there is a correlation

```python
#Create df_1 with only columns necessary for the plot
df_1 = df_order_values[['sum_order_value','income']]
```

```python
# Set plot size
plt.figure(figsize=(10, 6))  

#Define axis
sns.scatterplot(x='income', y='sum_order_value', data=df_1)

# Calculating and adding  R² value
slope, intercept = np.polyfit(df_1['income'], df_1['sum_order_value'], 1)
r_squared = np.corrcoef(df_1['income'], df_1['sum_order_value'])[0, 1]**2
plt.text(x=min(df_1['income']), y=max(df_1['sum_order_value']), s=f'R² = {r_squared:.2f}', fontsize=12)

# Setting title and axis
plt.title('Income and Order Value')
plt.xlabel('Income')
plt.ylabel('Order Value')

plt.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'income_order_value.png'))
```

With R² = 0.01 there is clearly no correlation

Calculate average order value for each customer over their lifetime so far

```python
# Calculate the average sum_order_value for each user_id
avg_order_value = df_order_values.groupby('user_id')['sum_order_value'].mean().reset_index()
```

```python
avg_order_value.columns = ['user_id', 'avg_order']
```

```python
avg_order_value
```

Create a histogram of order average order values to see how customer value is distributed

```python
plt.figure(figsize=(10, 6))  # Set the size of the figure
plt.hist(avg_order_value['avg_order'], bins=5, color='blue', alpha=0.7)

plt.title('Histogram of Average Order Value')
plt.xlabel('Average Order Value')
plt.ylabel('Number of Customers')

plt.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'histogram_order_value.png'))

plt.show()
```

```python
# Merge avg_order_value with df_order_values on 'user_id'
df_order_avg_values = df_order_values.merge(avg_order_value, on='user_id', how='left')
```

```python
df_order_avg_values.dtypes
```

```python
# Set avg_order to int16 to save memory
df_order_avg_values['avg_order'] = df_order_avg_values['avg_order'].astype('int16')
```

```python
df_order_avg_values.head()
```

```python
df_order_avg_values['avg_order'].describe()
```

Creating new variable cust_value to classify customer values based on avg_order

```python
df_order_avg_values.loc[df_order_avg_values['avg_order'] <= 81, 'cust_value'] = 'Lowest value'
df_order_avg_values.loc[(df_order_avg_values['avg_order'] > 81) & (df_order_avg_values['avg_order'] <113), 'cust_value'] = 'Low value' 
df_order_avg_values.loc[(df_order_avg_values['avg_order'] >= 113) & (df_order_avg_values['avg_order'] <153), 'cust_value'] = 'High value' 
df_order_avg_values.loc[df_order_avg_values['avg_order'] >= 153, 'cust_value'] = 'Highest value'
```

```python
#Creating frequency table for cust_value
df_order_avg_values['cust_value'].value_counts(dropna = False)
```

```python
#Looking at cust_value across regions
crosstab_cust_value = pd.crosstab(df_order_avg_values['cust_value'], df_order_avg_values['region'])
```

```python
crosstab_cust_value
```

Looking at the crosstab I can see that there won't be significat differences. However, I still want to visualize the data for the sake of excercise.

```python
# Group by region and cust_value and count occurrences
cust_value_counts = df_order_avg_values.groupby(['region', 'cust_value']).size().unstack(fill_value=0)
```

```python
cust_value_counts
```

```python
# Define custom sort order to ensure 'Lowwest value' comes first
sort_order = ["Lowest value", "Low value", "High value", "Highest value"]

# Reindex the DataFrame according to the custom sort order, filling missing columns with 0
sorted_cust_value_counts = cust_value_counts.reindex(columns=sort_order, fill_value=0)
```

```python
#Set plot size
plt.figure(figsize=(12, 8))

ax = sorted_cust_value_counts.plot(kind='bar', alpha=0.7)

# Adding titles and labels
plt.title('Number of Different Customers per Region', fontsize=15)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Number of Customers', fontsize=12)

# Move the legend to the right of the chart
plt.legend(title='Customer Value', bbox_to_anchor=(1.05, 1), loc='upper left')

# Export chart
export_path = os.path.join(path, '04 Analysis', 'Visualizations', 'bar_cust_value_region.png') 
plt.savefig(export_path, bbox_inches='tight') 

# Show plot
plt.xticks(rotation=45)
plt.tight_layout() 
plt.show()
```

Looking at customer value by age group

```python
# Define age bins and labels
bins = [-float('inf'), 25, 40, 50, 65, float('inf')]  # Define bins for age ranges
labels = ['<25', '25-40', '41-50', '51-65', '>65']  # Labels for those bins
```

```python
# Create a new column for age category
df_order_avg_values['age_group'] = pd.cut(df_order_avg_values['age'], bins=bins, labels=labels)
```

```python
# Group by age_group and cust_value and count occurrences
cust_value_counts_by_age = df_order_avg_values.groupby(['age_group', 'cust_value'], observed=False).size().unstack(fill_value=0)
```

```python
cust_value_counts_by_age
```

```python
# Define the custom sort order
sort_order = ["Lowest value", "Low value", "High value", "Highest value"]

# Reindex the DataFrame according to the custom sort order, filling missing columns with 0
sorted_cust_value_counts_by_age = cust_value_counts_by_age.reindex(columns=sort_order, fill_value=0)
```

```python
# Plot size
plt.figure(figsize=(12, 8))

ax = sorted_cust_value_counts_by_age.plot(kind='bar', alpha=0.7)

# Adding titles and labels
plt.title('Customers by Age Group', fontsize=15)
plt.xlabel(' ', fontsize=12) #deliberately left blank
plt.ylabel('Number of Customers', fontsize=12)

# Move the legend to the right of the chart
plt.legend(title='Customer Value', bbox_to_anchor=(1.05, 1), loc='upper left')

# Rotate labels for better visibility and adjust layout
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Export chart
export_path = os.path.join(path, '04 Analysis', 'Visualizations', 'bar_cust_value_age.png')
plt.savefig(export_path, bbox_inches='tight') 

# Show plot
plt.show()
```
